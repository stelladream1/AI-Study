```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
spamdata = pd.read_csv('/content/drive/MyDrive/spam.csv',encoding = 'ISO-8859-1')
```

```python
spamdata.head(5)
```

```python
spamdata.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
```

```python
spamdata.head(5)
```

```python
spamdata['v1'] = spamdata['v1'].map({'ham': 0, 'spam': 1})
```

```python
spamdata.head(5)
```

```python
spamdata.dropna()
```

```python
x = spamdata['v2']
y = spamdata['v1']
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x)
word_to_dic = tokenizer.word_index
print(len(word_to_dic))
```

```python
tokenizer = Tokenizer()
# tokenizer = Tokenizer( oov_token="<OOV>")
tokenizer.fit_on_texts(x) #문자를 입력받아서 리스트 형태로 변환해줌 
x = tokenizer.texts_to_sequences(x) #텍스트 단어들을 숫자로 변환 
```

```python
x_len = []
for i in x:
    x_len.append(len(i))
```

```python
max_len = max(x_len)
```

```python
x = pad_sequences(x, maxlen=max_len)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= .3, random_state = 1 , stratify = y)
```

```python

x_train, x_val, y_train, y_val = train_test_split(x_train,y_train, test_size= .3,random_state = 1)
```

```python
print(x_train.shape, x_val.shape, x_test.shape)
```

```python
print(y_train.shape, y_val.shape, y_test.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
vocab_size = len(word_to_dic) + 1
embedding_dim = 32
hidden_units = 32

model = keras.models.Sequential()
model.add(keras.layers.Embedding(vocab_size, embedding_dim))
model.add(keras.layers.SimpleRNN(hidden_units))
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(x_train, y_train, epochs=4, batch_size=32, validation_data=(x_val,y_val),verbose=1)
```

```python
model.summary()
```

```python
y_pred = model.predict(x_test)
```

```python
print(y_pred[:10])
```

```python
print(y_test[:10])
```

```python
y_pred =  y_pred.flatten()
```

```python
y_pred = np.where(y_pred > 0.5 ,1 , 0)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy : Training vs Validation')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc=0)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Loss : Training vs Validation')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc=0)
plt.show()
```
