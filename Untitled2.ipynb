```python
train = TensorDataset(x_train, y_train)
test = TensorDataset(x_test, y_test)
train_loader = DataLoader(train, batch_size=32, shuffle=True)
x_train, y_train = Variable(x_train), Variable(y_train)
```

```python
loss_fn = nn.CrossEntropyLoss()
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # 합성곱층
        self.conv1 = nn.Conv2d(3, 10, 5) # 입력 채널 수, 출력 채널 수, 필터 크기
        self.conv2 = nn.Conv2d(10, 20, 5)

        # 전결합층
        self.fc1 = nn.Linear(20 * 29 * 29, 50) # 29=(((((128-5)+1)/2)-5)+1)/2
        self.fc2 = nn.Linear(50, 2)

    def forward(self, x):
        # 풀링층
        x = F.max_pool2d(F.relu(self.conv1(x)), 2) # 풀링 영역 크기
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        x = x.view(-1, 20 * 29 * 29)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
```

```python
model = Net()
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
test_X = np.array(x_test, dtype='float32')
test_Y = np.array(y_test, dtype='int64')


test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()


print(test_X.shape)
print(test_Y.shape)


x_test, y_test = Variable(test_X), Variable(test_Y)
```

```python
loss_list = []
accuracy_list = []
count = 0
for epoch in range(100):
    total_loss = 0
    
    for x_train, y_train in train_loader:
        
        x_train, y_train = Variable(x_train), Variable(y_train)
        
        optimizer.zero_grad()
        
        h = model(x_train)
        
        loss = criterion(h, y_train)
        
        loss.backward()
        
        optimizer.step()
        
        total_loss += loss.data.item()

    result = torch.max(model(x_test).data, 1)[1]
    accuracy = sum(y_test.data.numpy() == result.numpy()) / len(y_test.data.numpy())
    print('[Epoch: {:>4}] cost = {:>.9}'.format(epoch + 1, total_loss))
    print(accuracy)
    loss_list.append(total_loss)
    accuracy_list.append(accuracy)
```
