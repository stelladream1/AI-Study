```python
class my_network(nn.Module):
    def __init__(self):
        super(my_network, self).__init__()
        self.conv1 = nn.Conv2d(3,64,5)
        self.conv2 = nn.Conv2d(64,32,5)
        self.fc1 = nn.Linear(32*5*5 , 128)
        self.fc2 = nn.Linear(128 , 10)
    def forward(self, x):
        x = F.relu(self.conv1(x), inplace = True)
        x = F.max_pool2d(x, (2,2))
  
        x = F.relu(self.conv2(x), inplace = True)
        x = F.max_pool2d(x, (2,2))
        
        x = x.view(x.shape[0], -1)
        #
        x = F.relu(self.fc1(x), inplace = True)
        x = F.relu(self.fc2(x), inplace = True)
        return x
```

```python
my_net = my_network()

optim = torch.optim.SGD(my_net.parameters(), lr=0.001, momentum = 0.9)
loss_function = nn.CrossEntropyLoss()
```

```python
num_epoch = 5

for epoch in range(num_epoch):
    for i, data in enumerate(trainloader):

        inputs, labels = data
        inputs,labels = Variable(inputs), Variable(labels)
        
        optim.zero_grad()
        out = my_net(inputs)

        loss = loss_function(out, labels)
        loss.backward()
        optim.step()

        if i % 64== 0:
            print("%d ===> loss : %.3f" %(i, loss))
print("over")
```

```python
total = 0
correct = 0
for data in test_loader:
    images , labes = data
    outputs = my_net(Variable(images))
    _, predicted = torch.max(outputs.data , 1)
    total = total + labels.size(0)
    correct = correct + (predicted == labels).sum()

print("%f" %(100 * correct / total))
```
