```python
train_path = "/content/drive/MyDrive/Gemstone/train" 
test_path = "/content/drive/MyDrive/Gemstone/test"  
```

```python
transform = transforms.Compose([ transforms.Resize((128, 128)),
                                transforms.ToTensor(),
                                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
                                ])
```

```python
trainset = torchvision.datasets.ImageFolder(root = train_path, 
                                            transform = transform)

testset = torchvision.datasets.ImageFolder(root = test_path,
                                            transform = transform)
```

```python
trainloader = DataLoader(trainset, batch_size = 64, shuffle = True, num_workers = 2,drop_last=True)
testloader = DataLoader(testset, shuffle = False, num_workers = 2,drop_last= False) # batch_size 안줘도됨 
```

```python
class my_network(nn.Module):
    def __init__(self):
        super(my_network, self).__init__()
        self.conv1 = nn.Conv2d(3,64,5)
        self.conv2 = nn.Conv2d(64,32,5)
        self.fc1 = nn.Linear(32 * 29 * 29 , 128)
        self.fc2 = nn.Linear(128 , 87)
    def forward(self, x):
        x = F.relu(self.conv1(x), inplace = True)
        x = F.max_pool2d(x, (2,2))
  
        x = F.relu(self.conv2(x), inplace = True)
        x = F.max_pool2d(x, (2,2))
        
        x = x.view(  64, -1 )
        #
        x = F.relu(self.fc1(x), inplace = True)
        x = F.relu(self.fc2(x), inplace = True)
        return x

model =my_network()
```

```python


optim = torch.optim.SGD(model.parameters(), lr=0.001, momentum = 0.9)
loss_function = nn.CrossEntropyLoss()
```

```python
num_epoch = 10
```

```python
for epoch in range(num_epoch):
    for i, data in enumerate(trainloader):

        inputs, labels = data
        inputs,labels = Variable(inputs), Variable(labels)
        optim.zero_grad()
        out = model(inputs)

        loss = loss_function(out, labels)
        loss.backward()
        optim.step()

        if  i% 64== 0:
            print("%d ===> loss : %.3f" %(epoch, loss))

            
print("over")
```

```python
total = 0
correct = 0
for data in testloader:
    images , labes = data
    outputs = model(Variable(images))
    _, predicted = torch.max(outputs.data , 1)
    total = total + labels.size(0)
    correct = correct + (predicted == labels).sum()

print("%f" %(100 * correct / total))
```
